```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(filename)
df
```

```python
# Simple Linear Regression
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
# Select the X [Predictor] and Y [Target Variable]

X = df[['highway-mpg']]
Y = df['price']
```

```python
# Splitting the data into training set and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
# Training the model
```

```python
lm.fit(X_train, Y_train)
```

```python
# Coefficient for each of the variables - Slope of the line
# Higher the coef, higher is the impact 
lm.coef_
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
# To find intercept of the line
lm.intercept_
```

```python
# Data Preparations
# Checking for null values
df.isna().sum()
```

```python
# Assign X[Predictors] and Y[Target Variable] variables to columns
X2 = df[["highway-mpg"]]
Y2 = df["price"]
```

```python
# Split for Training and Testing
from sklearn.model_selection import train_test_split
X_train2, X_test2, Y_train2, Y_test2 = train_test_split(X2,Y2, test_size=0.25, random_state = 1)
```

```python
# Scalling the data: Bringing the values in the df in the similar range
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train2)
X_test_scaler = scaler.transform(X_test2)
```

```python
# Model Development and evaluation 
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
```

```python
# Training the model [degree = 3]
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3) # degree decides overfitting, underfitting, or good fitting
X_poly_train = poly.fit_transform(X_train_scaler)
X_test_poly = poly.transform(X_test_scaler)
poly.fit(X_poly_train, Y_train2)
lin.fit(X_poly_train, Y_train2)
```

```python
# Prediction model [degree = 3]
Y_pred_test = lin.predict(X_test_poly)
# for evaluating the model
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y_test2, Y_pred_test)
```

```python
# Prediction [degree = 3]
# Analytical analysis [using training data]
Y_pred_train = lin.predict(X_poly_train)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y_train2, Y_pred_train)
```
